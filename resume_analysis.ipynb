```python
import PyPDF2

pdf_file_path="/content/sample_resume.pdf"
def extract_text_from_pdf(pdf_file_path):
    try:
        pdf_reader = PyPDF2.PdfReader(pdf_file_path)

        text = ''
        for page in pdf_reader.pages:
            text += page.extract_text()

        return text
    except Exception as e:
        print("An error occurred:", str(e))
        return None

# Replace 'example.pdf' with the path to your PDF file


extracted_text = extract_text_from_pdf(pdf_file_path)
if extracted_text:
    output_file_path = '/content/output.txt'  # Specify the path for the output text file
    with open(output_file_path, 'w', encoding='utf-8') as output_file:
        output_file.write(extracted_text)
    print("Text extracted from PDF and saved to", output_file_path)
```

```python
import zipfile
import pandas as pd
import io

# Specify the path to your ZIP file
zip_file_path = '/content/archive.zip'

# Create an empty DataFrame to store the merged data
merged_df = pd.DataFrame()

try:
    with zipfile.ZipFile(zip_file_path, 'r') as zip_file:
        # List all the CSV files in the ZIP archive
        csv_file_list = [name for name in zip_file.namelist() if name.endswith('.csv')]

        for csv_file_name in csv_file_list:
            with zip_file.open(csv_file_name) as file:
                # Read each CSV file within the ZIP archive into a DataFrame
                df = pd.read_csv(io.TextIOWrapper(file, encoding='utf-8'))

                # Append the data to the merged DataFrame
                merged_df = merged_df.append(df, ignore_index=True)

    # Specify the path where you want to save the merged CSV file
    merged_csv_path = '/content/merged_data.csv'

    # Save the merged data to a single CSV file
    merged_df.to_csv(merged_csv_path, index=False, encoding='utf-8')

    print("Merged CSV file saved to", merged_csv_path)

except Exception as e:
    print("An error occurred:", str(e))
```
